```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load cleaned data
df = pd.read_csv('../data/processed_tx_data.csv')
print("Data shape:", df.shape)
df.head()
```

```python
# Distribution of ETH Sent
plt.figure(figsize=(8, 5))
sns.histplot(df['Value_IN(ETH)'], bins=50, kde=True)
plt.title('Distribution of ETH Sent per Transaction')
plt.xlabel('Value_IN(ETH)')
plt.ylabel('Frequency')
plt.xscale('log')  # ETH values usually have a long tail
plt.show()
```

```python
# Distribution of Transaction Fees
plt.figure(figsize=(8, 5))
sns.histplot(df['TxnFee_USD'], bins=50, color='orange', kde=True)
plt.title('Distribution of Transaction Fees (USD)')
plt.xlabel('TxnFee_USD')
plt.ylabel('Frequency')
plt.xscale('log')
plt.show()
```

```python
# Hourly Activity Pattern
plt.figure(figsize=(10, 5))
sns.countplot(x='hour', data=df, palette='viridis')
plt.title('Ethereum Transactions by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Transaction Count')
plt.grid(True)
plt.show()
```

```python
# ETH Value vs Txn Fee (USD)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Value_IN_USD', y='TxnFee_USD', data=df, alpha=0.6)
plt.title('Txn Fee (USD) vs ETH Value (USD)')
plt.xlabel('Value_IN_USD')
plt.ylabel('TxnFee_USD')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()
```
